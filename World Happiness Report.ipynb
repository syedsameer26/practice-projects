```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/samir/Downloads/World Happiness Report 2022.csv")
df
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
plt.title("Null Values")
plt.show()
```

```python
sns.relplot(x = 'Happiness score', y = 'Explained by: GDP per capita' , data = df, hue = 'Country')
plt.figure(figsize = [15, 7])
plt.show()
```

```python
gdp_top = df.head().groupby('Country')['Happiness score'].sum()
gdp_top
```

```python
gdp_top.plot.pie(figsize = (20,10), fontsize = 10, autopct = '%.3f')
plt.show()
```

```python
gdp_b = df.tail().groupby('Country')['Happiness score'].sum()
gdp_b
```

```python
gdp_b.plot.pie(figsize = (20 , 10), fontsize = 10, autopct = '%.3f')
plt.show()
```
